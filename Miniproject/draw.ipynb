```python
import numpy as np
from skimage.draw import disk
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
from parameters import Acousticparameters

params = Acousticparameters()
n = params.nx
```

```python
img  = np.zeros((n,n), dtype=np.uint8)
center = n // 2 + n % 2
rr, cc = disk((center, center),5)
img[rr,cc] = 255

plt.imsave("./test_data/dot.png", img, cmap='gray')
plt.imshow(img)
```

```python
img  = np.zeros((n,n), dtype=np.uint8)
center = n // 2 + n % 2

thickness = 8
radius1 = 105
radius2 = 35

rr1, cc1 = disk((center, center),radius1 - thickness//2)
rr2, cc2 = disk((center, center), radius1 + thickness//2)
img[rr2,cc2] = 255
img[rr1,cc1] = 0

rr1, cc1 = disk((center, center), radius2 - thickness//2)
rr2, cc2 = disk((center, center), radius2 + thickness//2)
img[rr2,cc2] = 255
img[rr1,cc1] = 0

plt.imsave("./test_data/rings.png", img, cmap='gray')
plt.imshow(img)
```

```python
img  = np.zeros((n,n), dtype=np.uint8)

dx=28

radius = 1

center = n // 2 + n % 2

rr, cc = disk((center,center), radius)
img[rr,cc] = 255

rr, cc = disk((center+dx,center+dx), radius)
img[rr,cc] = 255

rr, cc = disk((center+dx,center-dx), radius)
img[rr,cc] = 255

rr, cc = disk((center-dx,center-dx), radius)
img[rr,cc] = 255

rr, cc = disk((center-dx,center+dx), radius)
img[rr,cc] = 255

plt.imsave("./test_data/dots.png", img, cmap='gray')
plt.imshow(img)
```
